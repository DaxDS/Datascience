```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'Year': [2017, 2018, 2019, 2020, 2021, 2022],
    'Vehicle_Type': ['Sedan', 'Sedan', 'SUV', 'SUV', 'Sedan', 'SUV'],
    'Sales': [150000, 160000, 170000, 140000, 180000, 190000]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Assuming you have information about recession years
recession_years = [2020, 2021]

# Set the Seaborn style
sns.set(style="whitegrid")

# Create the plot using Seaborn
plt.figure(figsize=(10, 6))

# Filter data for each vehicle type and plot lines
vehicle_types = df['Vehicle_Type'].unique()
for v_type in vehicle_types:
    v_type_data = df[df['Vehicle_Type'] == v_type]
    sns.lineplot(data=v_type_data, x='Year', y='Sales', label=v_type, marker='o')

# Highlight recession periods
for year in recession_years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)

# Shading for recession periods
ax = plt.gca()
for i in range(0, len(recession_years), 2):
    ax.fill_betweenx(ax.get_ylim(), recession_years[i], recession_years[i+1], color='gray', alpha=0.1)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Automobile Sales by Vehicle Type')

# Adding legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'Year': [2017, 2018, 2019, 2020, 2021, 2022],
    'Average_Price': [25000, 26000, 27000, 24000, 28000, 29000],
    'Sales': [150000, 160000, 170000, 140000, 180000, 190000]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Assuming you have information about recession years
recession_years = [2020, 2021]

# Create the scatter plot using Matplotlib
plt.figure(figsize=(10, 6))

# Scatter plot for non-recession years
non_recession_years = [year for year in df['Year'] if year not in recession_years]
plt.scatter(df[df['Year'].isin(non_recession_years)]['Average_Price'],
            df[df['Year'].isin(non_recession_years)]['Sales'],
            color='b', label='Non-Recession Years', alpha=0.7)

# Scatter plot for recession years
plt.scatter(df[df['Year'].isin(recession_years)]['Average_Price'],
            df[df['Year'].isin(recession_years)]['Sales'],
            color='r', label='Recession Years', alpha=0.7)

# Adding labels and title
plt.xlabel('Average Vehicle Price')
plt.ylabel('Sales Volume')
plt.title('Correlation between Average Vehicle Price and Sales Volume')

# Adding legend
plt.legend()

# Highlight recession years
for year in recession_years:
    plt.axvline(x=df[df['Year'] == year]['Average_Price'].values[0],
                color='gray', linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```
